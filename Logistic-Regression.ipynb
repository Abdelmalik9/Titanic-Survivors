# Libraries, directory and data

```python
#libraries 
import pandas as pd
import seaborn as sns
import numpy as np
import statsmodels.api as sm
import matplotlib.pyplot as plt
```

```python
#load the data
df = pd.read_csv("titanic.csv")
df.head()
```

# Data Analysis

```python
#histograms
df.hist(figsize= (15,10))
```

```python
#Drop name variable
df = df.drop(columns = "Name")
df.head(1)
```

```python
#Transform character variables
df = pd.get_dummies(data = df, drop_first = True,sparse=False,dtype=int)
```

```python
#Correlation matrix
plt.figure(figsize = (8,6))
sns.set(font_scale=1.5)
sns.heatmap(df.corr(),
            annot = True,
            fmt = '.1g',
            center = 0,
            cmap = 'coolwarm',
            linewidths = 1,
            linecolor = 'black')
```

# Logistic Regression preparation

```python
df.head(0)
```

```python
#Prepare X and Y
y = df.Survived
X = df.drop(columns = "Survived")
X = sm.add_constant(X)
```

```python
#Training and Test Split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y,
                                                    test_size = 0.2,
                                                    random_state = 1502)
X_train.head(2)
```

# Logistic Regression and Assessment

```python
#Logistic Regression
model = sm.Logit(y_train, X_train).fit()
print(model.summary())
```

```python
#predictions
predictions = model.predict(X_test)
predictions = np.where(predictions > 0.5, 1, 0)
predictions[:4]
```

```python
#Confusion matrix
from sklearn.metrics import confusion_matrix
print(confusion_matrix(y_test,predictions))
tn, fp, fn, tp = confusion_matrix(y_test, predictions).ravel()
```

```python
#Accuracy
#Accuracy = (True positive + True negative ) / All
accuracy = (tp + tn) / (tp + tn + fp + fn)
print("Accuracy:", round(accuracy * 100, 1), "%")

#F1-Score
#F1-score = 2 * TP / (2 * TP + FP + FN)
F1 = 2 * tp / (2 * tp + fp + fn)
print("F1-Score:" , round(F1 * 100, 1), "%")

#Specificity or True Negative Rate
#True Negative / (True Negative + False Positive)
Specificity = tn / (tn + fp)
print("Specificity:", round(Specificity * 100, 1), "%")

#Sensitivity, Recall or True Positive Rate
#True Positive / (True Positive + False Negative)
Sensitivity = tp / (tp + fn)
print("Sensitivity:", round(Sensitivity * 100, 1), "%")
```

```python
#Classification Report
from sklearn.metrics import classification_report
print(classification_report(y_test, predictions))
```
